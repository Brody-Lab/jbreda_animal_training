```python
import numpy as np
from scipy.stats import truncexpon
import matplotlib.pyplot as plt
```

## Resources

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.truncexpon.html
https://stackoverflow.com/questions/40143718/python-scipy-truncated-exponential-distribution


## Generate distribution

```python
LOWER = 1.5 # loc
UPPER = 5 # b
MEAN = 3 # scale

X = np.linspace(LOWER, UPPER, 10000)
ys = truncexpon.pdf(X, b=(UPPER-LOWER)/MEAN, loc=LOWER, scale=MEAN)
```

```python
plt.plot(X,ys)
```

## Generate Random Variables

```python
# to generate random variates x from trunc exp
random_var = truncexpon(b=(UPPER-LOWER)/MEAN, loc=LOWER, scale=MEAN)
samples = random_var.rvs(size=10000)

# note you need to norm them when plotting!!
fig, ax = plt.subplots(1, 1)
_ = ax.hist(samples, density=True, bins='auto', histtype='stepfilled', alpha=0.2, label="samples")
ax.plot(X, ys, label = "pdf", color='blue')
ax.set_title(f"Exponential with min: {LOWER}, mean: {MEAN}, max: {UPPER}")
plt.legend()
```
