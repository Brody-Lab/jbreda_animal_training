### Athena's growth

Based on looking at old data by athena her growth follows this logic:
* after non-violation trial, grow by 1% 
* if 1% is less than 1 ms, grow by 1 ms
* Do this until max (let's say 2s here)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trial_by_trial_growth(current_val, growth_fraction, minimum_growth) -> float:
    """
    This function calculates the growth of a value over time, given a growth fraction and a minimum growth value.
    :param current_val: The current value of the variable.
    :param growth_fraction: The fraction of growth.
    :param minimum_growth: The minimum growth value.
    :return: The new value of the variable.
    """

    growth_increment = current_val * growth_fraction
    if growth_increment < minimum_growth:
        growth_increment = minimum_growth
    new_val = current_val + growth_increment
    return new_val, growth_increment
```

```python
cur_delay = 0.001
all_delays = []
all_increments = []
max_delay = 2
while cur_delay < 2:
    cur_delay, growth_increment = trial_by_trial_growth(cur_delay, 0.001, 0.001)
    all_delays.append(cur_delay)
    all_increments.append(growth_increment)

all_increments = np.array(all_increments)
all_delays = np.array(all_delays)
```

```python
plt.subplot(2, 1, 1)
plt.plot(range(len(all_delays)), all_delays, label="delay")
plt.plot(
    range(len(all_delays)), all_increments, label="growth_increment", color="orange"
)
plt.legend()
plt.title(f"Grow Stats to {max_delay}s")
plt.xlabel("Trial")
plt.ylabel("Time [s]")

plt.subplot(2, 1, 2)
plt.plot(
    range(len(all_delays)), all_increments, label="growth_increment", color="orange"
)
plt.axhline(0.001, 0, len(all_delays), label="minimum_growth", color="black", ls="--")
plt.legend()
plt.ylabel("Growth Increment [s]")
plt.tight_layout()
```
